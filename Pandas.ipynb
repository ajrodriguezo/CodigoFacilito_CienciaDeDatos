### **Condicionales**

```python
# 1 Obtner el nombre de todos los usuarios con pais canada
# 2 Obtener el nombre y correo electronico de tos los usuarios con edad mayor a 50
# 3 Obtner el promedio de todos los usuarios de sexo femenino con una edad mayor a 30
```

```python
data = pd.read_csv('/content/users.csv',index_col='id')
data
```

```python
# limpiar datos
data=data.dropna()
data
```

```python
#1
data['pais']=='Canada'
```

### **Busqueda por rangos**

```python
# Obtnener todos los usuarios entre las edades 40 y 50
```

```python
data[(data['edad']>=40) & (data['edad']<=50)]
```

```python
data[data['edad'].between(40,50)]
```

### **B√∫squeda entre opciones**

```python
# obtner el nombre de los usuarios mayores a 30 a√±os
# de los paises Canada, alemania y francia
```

```python
data[data['edad']>30]
```

```python
data[(data['edad']>30) & ((data['pais']=='Canada')|(data['pais']=='Germany')|(data['pais']=='France'))]
```

```python
paises =['Canada','Germany','France']
```

```python
data[(data['edad']>30) & (data['pais'].isin(paises))]
```
